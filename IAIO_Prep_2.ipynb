```python
def col_by_yield(col, data):
    map_c = {}
    df_mean_train = data.groupby(col).mean()["yield"]

    index = 0;

    indexes = df_mean_train.index

    mat = df_mean_train.values

    for idx in indexes.values:
        map_c[idx] = mat[index]
        index += 1

    # data = pd.merge(data, df_mean_train, how="left", on=col)
    return map_c


def feature_engeneering(data):
    data["BlueberryCantuity"] = data["fruitmass"] / data["clonesize"]

    data["AverageTemp"] = (data["AverageOfUpperTRange"] + data["AverageOfLowerTRange"]) / 2

    data["AverageMaxTemp"] = (data["MaxOfUpperTRange"] + data["MinOfUpperTRange"]) / 2
    data["AverageMinTemp"] = (data["MaxOfLowerTRange"] + data["MinOfLowerTRange"]) / 2

    data["TempVariance"] = data["AverageOfUpperTRange"] - data["AverageOfLowerTRange"]

    data["TotalAverageBees"] = data["honeybee"] + data["bumbles"] + data["andrena"] + data["osmia"]

    data["AverageTemp/AverageRainingDays"] = data["AverageTemp"] / data["AverageRainingDays"]

    cols_drop = ["fruitmass", "clonesize", "AverageOfUpperTRange", "AverageOfLowerTRange", 
                 "MaxOfUpperTRange", "MinOfUpperTRange", "MaxOfLowerTRange", "MinOfLowerTRange",
                 "honeybee", "bumbles", "andrena", "osmia", "AverageRainingDays"]
    
    
    # data.drop(columns=cols_drop, inplace=True)

    return data

train_data = feature_engeneering(train_data)
test_data = feature_engeneering(test_data)

map_seeds = col_by_yield("seeds", train_data)
map_temps = col_by_yield("fruitmass", train_data)
map_rain = col_by_yield("RainingDays", train_data)

# train_data["yield_avg_seeds"] = train_data["seeds"].apply(lambda x: map_seeds[x])
# test_data["yield_avg_seeds"] = test_data["seeds"].apply(lambda x: map_seeds[x])


# train_data["yield_avg_fruitmass"] = train_data["fruitmass"].apply(lambda x: map_temps[x])
# test_data["yield_avg_fruitmass"] = test_data["fruitmass"].apply(lambda x: map_temps[x])

# train_data["yield_avg_rain"] = train_data["RainingDays"].apply(lambda x: map_rain[x])
# test_data["yield_avg_rain"] = test_data["RainingDays"].apply(lambda x: map_rain[x])

```

```python
from xgboost import XGBRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
# model = MLPRegressor(max_iter=100000, tol=1e-8, verbose=1)
model = RandomForestRegressor(n_estimators=1000, max_depth=8, criterion="absolute_error", n_jobs=-1, random_state=1)

model.fit(train_X, train_y)
```
