```python
import gymnasium as gym
from stable_baselines3 import PPO
import robot_env
```

```python
env = gym.make("chacha-robot", render_mode="no")
env.reset()

```

```python
model = PPO(env=env, policy="MlpPolicy")
model.save("PPO_Model_-1")
```

```python
for i in range(10):
    model.learn(1000)
    model.save("PPO_Model_" + str(i))
```

```python

env = gym.make('chacha-robot', render_mode='human')

model = PPO.load('PPO_Model_9', env=env)

obs = env.reset()[0]
terminated = False
while True:
    action, _ = model.predict(observation=obs, deterministic=True) # Turn on deterministic, so predict always returns the same behavior
    obs, _, terminated, _, _ = env.step(action)

    if terminated:
        break
```
