```python
yes_no_map = {
    "N": 0,
    "Y": 1
}

gender_map = {
    "M": 0,
    "F": 1
}

edema_map = {
    "N": 0, # No ??
    "S": 1, # Some ??
    "Y": 2, # Yes ??
}

status_map = {
    "C": 0,  # Alive
    "CL": 1, # Alive + Transplant
    "D": 2,  # Decesed
}

drug_map = {
    "Placebo": 1,
    "D-penicillamine": 2
}

def apply_encoding(data):
    # data['Drug'] = data['Drug'].map({'D-penicillamine': 0, 'Placebo':1})
    data['Sex'] = data['Sex'].map({'M': 0, 'F':1})
    data['Ascites'] = data['Ascites'].map({'N': 0, 'Y':1})
    data['Hepatomegaly'] = data['Hepatomegaly'].map({'N': 0, 'Y':1})
    data['Spiders'] = data['Spiders'].map({'N': 0, 'Y':1})
    data['Edema'] = data['Edema'].map({'N': 0, 'Y':1, 'S':2})
    if('Status' in data.columns): data['Status'] = data['Status'].map({'D': 2, 'C':0, 'CL':1})

    data["days_since_diagnosis"] = data["Age"] - data["N_Days"]
    # data["diseases"] = data['Ascites'] + data['Hepatomegaly'] + data['Spiders'] + data['Edema']
        
    # data["Age"] = data["Age"].apply(lambda x: x / 365)

    # df_means = data.mean()

    # data["Bilirubin"] = data["Bilirubin"].apply(lambda x: abs(x - df_means["Bilirubin"]))
    # data["Cholesterol"] = data["Cholesterol"].apply(lambda x: abs(x - df_means["Cholesterol"]))
    # data["Albumin"] = data["Albumin"].apply(lambda x: abs(x - df_means["Albumin"]))
    # data["Copper"] = data["Copper"].apply(lambda x: abs(x - df_means["Copper"]))
    # data["Alk_Phos"] = data["Alk_Phos"].apply(lambda x: abs(x - df_means["Alk_Phos"]))
    # data["SGOT"] = data["SGOT"].apply(lambda x: abs(x - df_means["SGOT"]))
    # data["Tryglicerides"] = data["Tryglicerides"].apply(lambda x: abs(x - df_means["Tryglicerides"]))
    # data["Platelets"] = data["Platelets"].apply(lambda x: abs(x - df_means["Platelets"]))
    # data["Prothrombin"] = data["Prothrombin"].apply(lambda x: abs(x - df_means["Prothrombin"]))

    

    data.drop(columns="Drug", inplace=True)

    return data

train_data = apply_encoding(train_data)
test_data = apply_encoding(test_data)
```
